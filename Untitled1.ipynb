```python
class Estudiante:
    def __init__(self, nombre, apellido, cedula, semestre):
        self.nombre = nombre
        self.apellido = apellido
        self.cedula = cedula
        self.semestre = semestre

    # ── Getters (Propiedades) ──────────────────────────────────
    @property
    def nombre(self): return self.__nombre

    @property
    def apellido(self): return self.__apellido

    @property
    def cedula(self): return self.__cedula

    @property
    def semestre(self): return self.__semestre

    # ── Setters con Validaciones ───────────────────────────────
    @nombre.setter
    def nombre(self, valor):
        self.__nombre = valor

    @apellido.setter
    def apellido(self, valor):
        self.__apellido = valor

    @cedula.setter
    def cedula(self, valor):
        # Validación: Solo números y exactamente 10 dígitos
        if valor.isdigit() and len(valor) == 10:
            self.__cedula = valor
        else:
            print(f"Error: La cédula '{valor}' debe ser numérica y tener 10 dígitos.")
            self.__cedula = "0000000000"  # Valor por defecto o manejo de error

    @semestre.setter
    def semestre(self, valor):
        # Validación: Rango del 1 al 9
        if 1 <= valor <= 9:
            self.__semestre = valor
        else:
            print(f"Error: El semestre {valor} es inválido (Rango permitido: 1-9).")
            self.__semestre = 1

    # ── __str__ ───────────────────────────────────────────────
    def __str__(self):
        return (f"Estudiante: {self.nombre} {self.apellido}\n"
                f"Cédula:     {self.cedula}\n"
                f"Semestre:   {self.semestre}°\n"
                f"{'-' * 35}")

# ── Pruebas de funcionamiento ────────────────────────────────
if __name__ == '__main__':
    print("--- Registro Exitoso ---")
    est1 = Estudiante("Maria", "Luna", "0923456781", 4)
    print(est1)

    print("--- Intento de Datos Erróneos ---")
    est2 = Estudiante("Carlos", "Reyes", "12345", 12) # Cédula corta y semestre fuera de rango
    print(est2)

    print("--- Corrección de Datos ---")
    est2.cedula = "0987654321" # Nueva cédula correcta
    est2.semestre = 9          # Nuevo semestre correcto
    print(est2)
```
