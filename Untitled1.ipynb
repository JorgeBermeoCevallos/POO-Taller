```python
class Estudiante:
    def __init__(self, nombre, apellido, cedula, semestre, email, materias=None):
        self.nombre = nombre
        self.apellido = apellido
        self.cedula = cedula
        self.semestre = semestre
        self.email = email
        self.materias = materias if materias is not None else []

    # ── Getters ────────────────────────────────────────────────
    @property
    def nombre(self): return self.__nombre

    @property
    def apellido(self): return self.__apellido

    @property
    def cedula(self): return self.__cedula

    @property
    def semestre(self): return self.__semestre

    @property
    def email(self): return self.__email

    @property
    def materias(self): return self.__materias

    # ── Setters ────────────────────────────────────────────────
    @nombre.setter
    def nombre(self, valor):
        self.__nombre = valor.strip().title()

    @apellido.setter
    def apellido(self, valor):
        self.__apellido = valor.strip().title()

    @cedula.setter
    def cedula(self, valor):
        if valor.isdigit() and len(valor) == 10:
            self.__cedula = valor
        else:
            print(f"Error: La cedula '{valor}' debe tener 10 digitos.")
            self.__cedula = "0000000000"

    @semestre.setter
    def semestre(self, valor):
        if 1 <= valor <= 9:
            self.__semestre = valor
        else:
            print(f"Error: Semestre {valor} invalido (Rango 1-9).")
            self.__semestre = 1

    @email.setter
    def email(self, valor):
        if "@" in valor and "." in valor:
            self.__email = valor.lower().strip()
        else:
            print(f"Error: Email '{valor}' no valido.")
            self.__email = "pendiente@correo.com"

    @materias.setter
    def materias(self, valor):
        if isinstance(valor, list):
            self.__materias = valor
        else:
            print("Error: Materias debe ser una lista.")
            self.__materias = []

    # ── __str__ ───────────────────────────────────────────────
    def __str__(self):
        mats_texto = ", ".join(self.materias) if self.materias else "Sin materias"
        return (f"Estudiante: {self.nombre} {self.apellido}\n"
                f"Cedula:     {self.cedula}\n"
                f"Semestre:   {self.semestre}\n"
                f"Email:      {self.email}\n"
                f"Materias:   {mats_texto}\n"
                f"{'-' * 35}")

# ── Ejecución Directa ────────────────────────────────────────
if __name__ == '__main__':
    # Estudiante 1: Datos correctos
    est1 = Estudiante(
        nombre="ana lucia",
        apellido="vargas",
        cedula="0912345678",
        semestre=3,
        email="ana.vargas@u.edu.ec",
        materias=["Programacion", "Calculo", "Fisica"]
    )
    print(est1)

    # Estudiante 2: Probando validaciones automáticas
    est2 = Estudiante(
        nombre="luis",
        apellido="mora",
        cedula="123",        # Cedula incorrecta
        semestre=15,         # Semestre fuera de rango
        email="luismora.com", # Email sin @
        materias="Ninguna"    # No es una lista
    )

    # Corrigiendo mediante setters
    est2.cedula = "0987654321"
    est2.semestre = 5
    est2.email = "luis.mora@u.edu.ec"
    est2.materias = ["Base de Datos", "Redes"]

    print(est2)
```
